As you can see, our data doesn't have homogenous mean and standard deviation, the range (min-max also varies!)
Many ML models using distance to inform them, so feature with larger scale may have more influence than the other.
We want to "Normalize" - scaling and centering the data 

Several methods can be used such as [StandardScale](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) which you used earlier; or [MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html#sklearn.preprocessing.MinMaxScaler).
StandardScaler bringing all features' mean to 0 and std to 1, while MinMaxScaler substract the min and divide entry data by the range.

```X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
X_scaled = X_std * (max - min) + min```

````python
# Import scale
from sklearn.preprocessing import scale

# Scale the features: X_scaled
data_scaled = scale(data)

# Print the mean and standard deviation of the unscaled features
print("Mean of Unscaled Features: {}".format(np.mean(data))) 
print("Standard Deviation of Unscaled Features: {}".format(np.std(data)))

# Print the mean and standard deviation of the scaled features
print("Mean of Scaled Features: {}".format(np.mean(data_scaled))) 
print("Standard Deviation of Scaled Features: {}".format(np.std(data_scaled)))
````

## Pipeline


We can chain all the preprocessing (scaling, impute missing data...), model instantiation and cross validation in 1 pipeline!

````python
# Import the Imputer module
from sklearn.impute import SimpleImputer 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Create scaler: scaler
scaler = StandardScaler()
# Setup the Imputation transformer: imp
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Instantiate the SVC classifier: clf
pca = PCA(n_components=402)

# Setup the pipeline with the required steps: steps
steps = [('imputation', imp), 
         ('scaler', scaler),
        ('pca', pca)]
# Create the pipeline: pipeline
pipeline = Pipeline(steps)

X = data
y = meta.disease_staging.values.reshape(-1,1)


# Fit the pipeline 
pipeline.fit(X,y_train)

# Transform
pca_features = pipeline.transform(X)
````

##### Class imbalance

    Consider this situation:
        Your model predict survival of cancer patients. Most of the time the patients die (let's be negative, 90%).
        You build a model predict ALL patients die.
        Your model accuracy is 90% accurate!
        But horrible at classifying those patients will survive! We need another more nuanced metric 

__Confusion matrix__

## Hyperparameter tuning

We see that all we do for linear regression

````python
# Import necessary modules
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Setup the hyperparameter grid
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}

# Instantiate a logistic regression classifier: logreg
logreg = LogisticRegression()

# Instantiate the GridSearchCV object: logreg_cv
logreg_cv = GridSearchCV(logreg, param_grid, cv=5)

# Fit it to the data
logreg_cv.fit(brcc_df,brcc_tg)

# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_)) 
print("Best score is {}".format(logreg_cv.best_score_))
````
