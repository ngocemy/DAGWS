## Centering and scaling

````python
import pandas as pd
data = pd.read_csv("data/Combat_filtered_exprs.txt",delimiter="\t")
data = data.transpose()
data = data.sort_index()
data.head()
````

````python
data.describe()
````

As you can see, our data doesn't have homogenous mean and standard deviation, the range (min-max also varies!)
Many ML models using distance to inform them, so feature with larger scale may have more influence than the other.
We want to "Normalize" - scaling and centering the data 

Several methods can be used such as [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) which you used earlier; or [MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html#sklearn.preprocessing.MinMaxScaler).
StandardScaler bringing all features' mean to 0 and std to 1, while MinMaxScaler substract the min and divide entry data by the range.

```X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
X_scaled = X_std * (max - min) + min```

````python
# Import scale
from sklearn.preprocessing import scale
import numpy as np
````

````python
# Scale the features: X_scaled
data_scaled = scale(data)
````

````python
# Print the mean and standard deviation of the unscaled features
print("Mean of Unscaled Features: {}".format(np.mean(data))) 
print("Standard Deviation of Unscaled Features: {}".format(np.std(data)))
````

````python
# Print the mean and standard deviation of the scaled features
print("Mean of Scaled Features: {}".format(np.mean(data_scaled))) 
print("Standard Deviation of Scaled Features: {}".format(np.std(data_scaled)))
````

# Model assessment & Hyperparameter tuning

    We calculated once the accuracy of our model prediction. Do you think it's always a good metric to assess our model?

##### Class imbalance

    Consider this situation:
        Your model predict survival of cancer patients. Most of the time the patients die (let's be negative, 90%).
        You build a model predict ALL patients die.
        Your model accuracy is 90% accurate!
        But horrible at classifying those patients will survive! We need another more nuanced metric 

__Confusion matrix__

<img src="fig/confusion.png" width=400>

    Class of interest is called the “positive”, we want to predict patient survive, then live is positive. (However it really depends on your convention)

- Other metrics:

            - Metrics for classification

<img src="fig/metrics.png" width=300>

````python
from sklearn.datasets import load_breast_cancer
brcc = load_breast_cancer()
import pandas as pd
brcc_df = pd.DataFrame(brcc.data,columns=brcc.feature_names)
brcc_tg = pd.Series(brcc.target,name='Target',dtype='category')
brcc_df.head()
````

````python
# Import necessary modules
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  scale
````

````python
brcc_scaled = scale(brcc_df)
````

````python
# Create training and test set
X_train, X_test, y_train, y_test = train_test_split(brcc_scaled,brcc_tg,test_size=0.4,random_state=42)
````

````python
# Instantiate a k-NN classifier: knn
knn = KNeighborsClassifier(n_neighbors=6)
````

````python
# Fit the classifier to the training data
knn.fit(X_train,y_train)
````

````python
# Predict the labels of the test data: y_pred
y_pred = knn.predict(X_test)
````

````python
knn.score(X_train,y_train)
````

````python
knn.score(X_test,y_test)
````

````python
# Generate the confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
````

### ROC curve

<img src="logistic_reg.png" width=300>

````python
# Import the necessary modules
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
````

````python
# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(brcc_scaled,brcc_tg, test_size = 0.4, random_state=42)
````

````python
# Create the classifier: logreg
logreg = LogisticRegression()
````

````python
# Fit the classifier to the training data
logreg.fit(X_train,y_train)
````

````python
# Predict the labels of the test set: y_pred
y_pred = logreg.predict(X_test)
````

````python
# Compute and print the confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
````

````python
# Ploting a ROC curve
````

````python
# Import necessary modules
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
````

````python
# Compute predicted probabilities: y_pred_prob
y_pred_prob = logreg.predict_proba(X_test)[:,1]
````

````python
# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
````

````python
# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
````

````python
# Import necessary modules
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score

# Compute and print AUC score
print("AUC: {}".format(roc_auc_score(y_test, y_pred_prob)))

# Compute cross-validated AUC scores: cv_auc
cv_auc = cross_val_score(logreg,brcc_scaled,brcc_tg, cv=5,scoring='roc_auc')

# Print list of AUC scores
print("AUC scores computed using 5-fold cross-validation: {}".format(cv_auc))
````

## Hyperparameter tuning

We see that all we do for linear regression is finding the 
Linear regression: y = ax + b; a and b are called parameters because after fitting the model learns them.

LASSO (regularized) regression: choosing alpha

k-nearest neighbor (classification): choosing k 

`Alpha` and `k` here is called __Hyperparameters__ because it can't be learned by fitting the model

One of the approach to tune a set of hyperparameters is using GridSearch.
We perform k-fold cross validation for each combination of hyperparameters, and then choosing which one performs the best!
Let's try GridSearch on the KNN above example!

````python
# Import necessary modules
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
````

````python
# Setup the hyperparameter grid
n_neighbors_space = np.linspace(2, 8, 30,dtype=int)
param_grid = {'n_neighbors': n_neighbors_space}
````

````python
# Instantiate a logistic regression classifier: logreg
knn = KNeighborsClassifier()
````

````python
# Instantiate the GridSearchCV object: logreg_cv
knn_cv = GridSearchCV(knn, param_grid, cv=5)
````

````python
# Fit it to the data
knn_cv.fit(brcc_df,brcc_tg)
````

````python
# Print the tuned parameters and score
print("Tuned KNN Parameters: {}".format(knn_cv.best_params_)) 
print("Best score is {}".format(knn_cv.best_score_))
````

## Pipeline


We can chain all the preprocessing scaling, model instantiation and cross validation in 1 pipeline!

````python
# Import the Pipeline module
from sklearn.pipeline import Pipeline
````

````python
# Setup the pipeline
steps = [('scaler', StandardScaler()), # Here I use a new scaler, see what is the difference?
         ('KNN', KNeighborsClassifier())]
````

````python
pipeline = Pipeline(steps)
````

````python
np.linspace(2, 8, 30,dtype=int)
````

````python
# Specify the hyperparameter space
parameters = {'KNN__n_neighbors':np.linspace(2, 8, 30,dtype=int)}
# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(brcc_df,brcc_tg,test_size=0.2,random_state=21)

# Instantiate the GridSearchCV object: cv
cv = GridSearchCV(pipeline,parameters,cv=3)

# Fit to the training set
cv.fit(X_train,y_train)

# Predict the labels of the test set: y_pred
y_pred = cv.predict(X_test)

# Compute and print metrics
print("Accuracy: {}".format(cv.score(X_test, y_test)))
print(classification_report(y_test, y_pred))
print("Tuned Model Parameters: {}".format(cv.best_params_))
````
